```python
import datetime 
start=datetime.datetime.now() 

def karsilastir(girilen,text):
    bulunandizi=[]
    strb=text #gelen veri strb degiskenine atılıyor
    test=girilen in strb #girilen deger strb nin içinde varsa test True döner Yoksa False döner
    if test==True:
        print("şunu mu demek istediniz:",text)
        bulunandizi.append(text) #bos diziye text değeri ekleniyor
    else:
        pass
    return bulunandizi


giris=input("Aranan harfler") #kullanıcıdan aranacak veri alınır
f=open("deneme.txt","r",encoding="utf-8") #deneme.txt ye bağlantı kurulur
kelimeler=f.readlines(); #kelimeler dizisine okunan değerler atanır
f.close()
a=[]
f=open("bulunan.txt","w",encoding="utf-8")#bulunan.txt ye bağlantı sağlanır
for i in kelimeler: #kelimeler içinde bir döngü oluşturulur
    a.append(karsilastir(giris.lower(),i.lower()))
for i in range(0,len(a)): #a nın uzunluğu boyunca dönen bir döngü oluşturulur..Son elemana kadar dön.
    if len(a[i])>0: #eğer a nın i. değeri boş değilse
        f.write(a[i][0])
stop=datetime.datetime.now()
zaman=stop-start
print("Yürütme zamanı : ",zaman,"saniye")
f.close()
        
```
